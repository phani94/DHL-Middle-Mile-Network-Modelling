```python
import numpy as np
import pandas as pd

#read_file_Costs = pd.read_csv('Database/Costs.csv')
input_costs = pd.read_excel('Database/Input_costs_from_current_DHL.xlsx')
```

```python
read_file_Costs = pd.read_csv('Database/Costs.csv')
read_file_time= pd.read_csv('Database/travel time.csv')
n=read_file_Costs.columns
n
```

```python
# Merging Costs and Travel time csv files
df_c=read_file_Costs
df_t=read_file_time

df_c['hub']=df_c['From']+'-'+df_c['To']

df_t['hub']=df_t['From']+'-'+df_t['To']

'''
n=df_t.columns
n=n.to_list()
n.remove('hub')
n.remove('est_tt')
n.remove('Duration')

df = df.drop(df.columns.difference(n), axis=1)
'''
merged_df= pd.merge(df_c, df_t, on='hub', how='left')

df_c['Duration']=merged_df['Duration'].to_list()
df_c['est_tt']=merged_df['est_tt'].to_list()
df_c
```

```python
# To convert city to Ref_city

hub_list = pd.read_excel('Database/Hub list.xlsx', encoding='unicode_escape')
hub_list.Ref_City = hub_list.Ref_City + "-" + hub_list.Country
hub_ref = {}
for city in hub_list.City.unique():
    hub_ref[city] = hub_list[hub_list["City"] == city]["Ref_City"].tolist()[0]

def name_transform(city_name):
    if city_name in hub_ref.keys():
        #print(city_name, " : ", hub_ref[city_name])
        return hub_ref[city_name]
    else:
        print(city_name, " not Found")
        return city_name

```

```python
# DHL current Input data with costs..
input_costs

len(input_costs)
```

```python
# Arranging data and truck types into 4 main types in DHL current Input data.
print(len(input_costs))
input_costs.replace("trailer", "Trailer", inplace = True)
input_costs.replace("2 swap bodies", "2 Swap Bodies", inplace = True)
input_costs.replace("van", "Van", inplace = True)
input_costs.replace("1 swap body", "Swap Body", inplace = True)
input_costs.replace("SEMI TRAILER", "Swap Body", inplace = True)
input_costs.replace("Cargo 50 m3", "Swap Body", inplace = True)
input_costs.replace("Sea Container", "2 Swap Bodies", inplace = True)
input_costs.replace("Van 15 m3", "Van", inplace = True)
input_costs.replace("Van 22m3", "Van", inplace = True)
input_costs.replace("4 swap bodies", "2 Swap Bodies", inplace = True)
#df=input_costs.dropna()
#input_costs.dropna(inplace=True)
input_costs.dropna(subset = ["Truck type"], inplace=True)
df=input_costs
df = df[~df['Distance in km'].isin(['Multistop'])]

pd.set_option('display.max_rows', df.shape[0]+1)
print(len(df))
df
```

```python
p=[]

q=[]
for i in df['Org. GTW']:
    p.append(name_transform(i))
    
for i in df['Dst. GTW']:
    q.append(name_transform(i))
    
    
df['From']=p
df['To']=q

df['hubs']= df['From']+'-'+df['To']+'-'+df['Truck type']
#pd.set_option('display.max_rows', df.shape[0]+1)
print(len(df))
df
```

```python
# There are the costs what we predicted before ..
costs=df_c
n=costs.columns
n
```

```python
costs['hubs']=costs['From']+'-'+costs['To']+'-'+costs['Truck type']
#pd.set_option('display.max_rows', cos.shape[0]+1)
#costs=costs[costs['hubs'] != np.nan]
#costs.dropna(subset = ["hubs"], inplace=True)
len(costs)

```

```python
# Mergin DHL current Input and previously modelled costs data 

merged_df= pd.merge(costs, df, on='hubs', how='left')
#merged_df=pd.merge(left=costs, right=df, how='inner', left_on='hubs_p', right_on='hubs_q')
m=merged_df
u=merged_df
n=merged_df.columns
print(n)
merged_df
```

```python
# Here we remove previous costs and other unnecessary data from the above merged data..

merged_df=merged_df.drop(['Unnamed: 0','From_x', 'To_x', 'Truck type_x', 'Ori. ','Dest.', 'Org. GTW', 'Dst. GTW','Truck type_y','From_y','To_y', 'Costs_x','Capacity','Cost/kg','Distance in km'],1)

#Replace all junk values with NaN

v=merged_df['Costs_y']
v=v.replace('?', np.NaN)
v=v.replace('n.a.', np.NaN)
v=v.replace(0, np.NaN)
v=v.replace(r'^\s*$', np.NaN, regex=True)
merged_df['Costs']=v.to_list()
merged_df.to_excel('Comparision/merged.xlsx')
len(merged_df)
```

```python
# Data proccessed for modeling 
proccessed_data=merged_df.drop(['hubs','Duration','hub','Costs_y'],1)
proccessed_data

```

```python
# Apply LR to impute missing costs
from sklearn.linear_model import LinearRegression
linreg= LinearRegression()

# Costs with null/Nan
data_with_null=proccessed_data

# Costs without null/Nan
data_without_null=proccessed_data.dropna()

# training data
train_data_x=data_without_null.iloc[:, :2]
train_data_y=data_without_null.iloc[:, 2]

linreg.fit(train_data_x,train_data_y)


```

```python
#testing data
test_data = data_with_null.iloc[:,:2]

# only missing costs
proccessed_data['Costs']=pd.DataFrame(linreg.predict(test_data))

# Commbing missing costs with existing costs
data_with_null.Costs.fillna(proccessed_data.Costs, inplace=True)
data_with_null.to_csv('Comparision/LRmodel.csv')
data_with_null
```
