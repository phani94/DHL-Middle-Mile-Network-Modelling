```python
import numpy as np
import pandas as pd


# Currents demands given by DHL, divided by demands
low_demands = pd.read_csv('Comparision/low_Demand_Co_loading.csv') # First Ten  in DHL given list
med_demands = pd.read_csv('Comparision/Medium_Demand_Co_loading.csv')  # 10-41
high_demands = pd.read_csv('Comparision/High_Demand_Co_loading.csv') #41-70

# These are the files which combine costs, distances, travel times and outputs of Recommended model

rec_output = pd.read_csv('Comparision/Recommended_Output.csv') 






  
```

```python
def removeUnwanted_cols_list(df, ls):
    for i in ls:
        
        cols = [c for c in df.columns if i not in str(c)]

        df=df[cols]
    return df
```

```python

# Merginging Recommended Outputs with Current model Outputs, based on their demand level..

def convert_To_Countries(df, Rec):
    
    new = df["Origin_Country_DHL"].str.rsplit("-", n = 1, expand = True) 
    df["Origin_Country_DHL"]= new[0] 
    new = df["Destination_Country_DHL"].str.rsplit("-", n = 1, expand = True) 
    df["Destination_Country_DHL"]= new[0]

    df['Countries']=df['Origin_Country_DHL']+"-"+df['Destination_Country_DHL']
    #print(df)
    df['Monthly_DHL']= df['Weekly_DHL']*4
    
    
    
    Rec['Countries']=Rec['Origin Country']+"-"+Rec['Destination Country']
    
    Test_results= pd.merge(df, Rec, on='Countries', how='inner')
    
    
    ls=['Countries','Origin_Country_DHL','Destination_Country_DHL', 'Unnamed']
    
    Test_results=removeUnwanted_cols_list(Test_results, ls)
    #club=club.drop(['Countries','Origin_Country_DHL','Destination_Country_DHL'], 1)
    
    
    return Test_results



low_Demand_Test=convert_To_Countries(low_demands, rec_output)

Medium_Demand_Test=convert_To_Countries(med_demands, rec_output)

High_Demand_Test=convert_To_Countries(high_demands, rec_output)





  
```
