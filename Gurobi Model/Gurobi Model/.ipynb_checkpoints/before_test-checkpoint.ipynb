```python
def removeUnnamed(df, ss):
    cols = [c for c in df.columns if ss not in str(c)]

    df=df[cols]
    return df
    

        
    


```

```python
#Add column name to Frequency file

read_file_freq=removeUnnamed(read_file_freq, 'Unnamed')
read_file_freq.columns=['From', 'To', 'Truck type', 'Frequency']

freq=read_file_freq[read_file_freq['Frequency']>0]
freq

```

```python
#Add coulmn names to Route Capacity file
read_file_route=removeUnnamed(read_file_route, 'Unnamed')


read_file_route.columns=['From', 'To',  'Route Capacity']
read_file_route

r_cap=read_file_route[read_file_route['Route Capacity']>0]
len(r_cap)
```

```python
def combine_hubs(X, Y, Z):
    
    if(Z==0):
        X["hubs"]= X['From']+"-"+X['To']
        Y["hubs"]= Y['From']+"-"+Y['To']
        merged_df= pd.merge(X, Y, on='hubs', how='inner')
        return merged_df
    else:
        X["hubs_TT"]= X['From']+"-"+X['To']+X['Truck type']
        Y["hubs_TT"]= Y['From']+"-"+Y['To']+Y['Truck type']
        merged_df= pd.merge(X, Y, on='hubs_TT', how='inner')
        return merged_df
    
    

Freq_Rcap_output=combine_hubs(freq,r_cap,0)
Freq_Rcap_output=removed_unwanted_columns(Freq_Rcap_output)



Costs_Merge_with_output=combine_hubs(Freq_Rcap_output,costs,1)
Costs_Merge_with_output=removed_unwanted_columns(Costs_Merge_with_output)

travel_time_Merge_with_output=combine_hubs(Costs_Merge_with_output,travel_time,0)
travel_time_Merge_with_output=removed_unwanted_columns(travel_time_Merge_with_output)

travel_time_Merge_with_output.to_csv('Comparision/Recommended_Output.csv')
travel_time_Merge_with_output

```
